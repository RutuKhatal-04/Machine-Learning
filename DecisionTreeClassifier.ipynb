```python
import pandas as pd
```

```python
data=pd.read_csv('/content/iris.csv')
```

```python
data=data['species'].replace({'setosa':0,'virginica':1,'versicolor':2})
data
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
x=data[['sepal_length']]
y=data['species']

clf = tree.DecisionTreeClassifier()
clf.fit(x,y)

y=clf.predict(x)
```
