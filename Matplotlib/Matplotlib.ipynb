```python
import pandas as pd
```

```python
data=pd.read_csv("/content/sample_data/california_housing_test.csv")
data
```

```python
X=data.households.values
Y=data.population.values
```

```python
print(type(X))
print(type(Y))
```

```python
fig,ax=plt.subplots()
ax.bar(X,Y,2,color='green')
ax.plot(X,Y,color='red')
ax.set_title("Bar Plot")
ax.set_xlabel("households")
ax.set_ylabel("population")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X=data[['households']]
Y=data[['population']]
```

```python
model.fit(X,Y)
```

```python
fig,ax=plt.subplots()
ax.plot(X,X*model.coef_,2,color='green') #(x,m*x) here m= model.coef_

ax.set_title("Bar Plot")
ax.set_xlabel("households")
ax.set_ylabel("population")
# ax.grid(true)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
X=data[['households']] # Keep the double brackets to create a DataFrame
Y=data.population
model1.fit(X,Y.values)
fig,ax=plt.subplots()
ax.plot(X,X*model1.coef_.reshape(-1,1),color='blue')
ax.plot(X,model.predict(X),color='red')
ax.set_title("Bar Plot")
ax.set_xlabel("households")
ax.set_ylabel("population")
plt.show()
```
