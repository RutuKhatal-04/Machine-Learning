# Working On Txt


```python
import pandas as pd
university_towns=[]
with open('/content/university_towns.txt') as file:
  for line in file:
    if '[edit]' in line:
      #Remember this 'state' untill the next is found
      state=line
    else:
      #Otherwise we have a city , keep 'state' as last-seen
      university_towns.append((state,line))
university_towns
```

```python
#Converting List Into Pandas
towns_df=pd.DataFrame(university_towns,columns=['State','RegionName'])
towns_df
towns_df['State']=towns_df['State'].str.replace('[edit]\n','')
towns_df
towns_df['RegionName']=towns_df['RegionName'].str.replace('\n','')
towns_df
```

```python
def get_cityname(item):
  if '[' in item:
    return item[:item.find('[')]
  # elif '(' in item:
  #   return item[:item.find('(')]
  else:
    return item
```

```python
towns_df=towns_df.applymap(get_cityname)
towns_df
```

```python
towns_df=towns_df.applymap(get_cityname)
towns_df
def get_city(item):
  if ('(' in item):
    return item[item.find('(')+1:item.find(')')]
  else:
    return item
towns_df=towns_df.applymap(get_city)
towns_df
```
