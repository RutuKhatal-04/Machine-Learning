```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = datasets.load_iris()
fd = pd.DataFrame(data=iris.data, columns=iris.feature_names)
fd['species'] = iris.target

# Select the feature and the target
x = fd[['sepal length (cm)']]
y = fd[['species']]

# Convert to binary classification (Iris-setosa vs. others)
y = (y != 0) * 1

# Split the data into training and testing sets
x_train = x[:-20]
x_test = x[-20:]
y_train = y[:-20]
y_test = y[-20:]

# Train the logistic regression model
model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(x_train, y_train)

# Plotting
plt.figure(figsize=(10, 6))
plt.title('Logistic Regression on Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Species (0: Setosa, 1: Non-setosa)')

# Scatter plot for training and testing data
plt.scatter(x_train, y_train, color='green', label='Train')
plt.scatter(x_test, y_test, color='red', label='Test')

# Plot the decision boundary
x_values = np.linspace(x.min(), x.max(), 100)
y_values = model.predict_proba(x_values)[:, 1]
plt.plot(x_values, y_values, color='blue', label='Decision Boundary')

plt.legend()
plt.grid(True)
plt.show()
```
