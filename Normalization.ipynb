#Min-max Scaler

Transform features by scaling each feature to given range. This estimator scales and translate each feature  individually such that it is in the given range on the training set, e.g. between zero and one.

```python
from sklearn.preprocessing import minmax_scale
X=[[-2,1,2],[-1,0,1]]
minmax_scale(X,axis=0) #Scales each column independently
minmax_scale(X,axis=1) #Scales each row independently
```
