```python
#Univariate
#Hypothesis testing
#sampling
import scipy.stats
mu=0                        #mean
variance = 2                #variance
sigma=np.sqrt(variance) #standard deviation
x=np.linspace(mu-3*variance,mu+3*variance,100) #kurtosis value is (-3 - 3)

plt.plot(x,scipy.stats.norm.pdf(x,mu,sigma)) #nomr is normal distribution function
#pdf is probablity density function
```

```python
fvalues=np.linspace(.1,5,100)

#pdf(x,df1,df2): Probablity density function at x of F
plt.plot(fvalues,scipy.stats.f.pdf(fvalues,1,30),'b-',label="F(1,30)")
plt.plot(fvalues,scipy.stats.f.pdf(fvalues,5,30),'r-',label="F(5,30)")
plt.legend()
```
